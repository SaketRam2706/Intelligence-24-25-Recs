Data preparation

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
import os.path

import joblib
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.pipeline import Pipeline
import time
from datetime import datetime as dt
from string import punctuation
import nltk
from nltk.corpus import stopwords
```

```python
label_mapping = []
```

```python
def read_dataset(data):
    df = data.copy()  

    if(df.shape[1]==4):
        df['label'] = label_encoder.fit_transform(df['Category'])
        label_mapping.append(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
        df = df.rename(columns={"encoded": "label", "News_headline": "text", "News_title" : "title"})
    else:
        df = df.rename(columns={"News_headline": "text", "News_title" : "title"})

    initial_shape = df.shape[0]
    df = df.dropna(how="any")
    curr_shape = df.shape[0]
    print(f"There were {initial_shape - curr_shape} empty examples dropped.")

    df.loc[:, "full_text"] = df["title"] + " " + df["text"]
    
    return df
```

```python
df_train = read_dataset(df)
df_train
```

```python
df_test["full_text"] = df_test["News_title"] + " " + df_test["News_headline"].fillna("")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Hyperparameters
MAX_DF = 0.95
MIN_DF = 15
N_JOBS = -1   


pipeline = Pipeline(steps=[
    ("vectorizer", TfidfVectorizer(max_df=MAX_DF,
                                    min_df=MIN_DF,
                                    stop_words='english',  
                                    token_pattern=r'(?ui)\b\w*[a-zA-Z]+\w*\b')), 
    ("naive_bayes", MultinomialNB())
], verbose=True)


pipeline.fit(df_train["full_text"], df_train["label"])
```

```python
predictions = pipeline.predict(df_test["full_text"])
df_test["predicted_label"] = predictions
df_test["preicted_categories"] = label_encoder.inverse_transform(predictions)
df_test
```

```python
df_selected = df_test[['ID', 'predicted_label']]

df_selected.to_csv('new_file_2.csv', index=False)
```
